```python
from pyspark.sql.functions import *
```

```python
df = spark.read.format("parquet")\
                .option("header", "true")\
                .option("inferschema", "true")\
                .load("abfss://bronze@olympicprojdlchintu.dfs.core.windows.net/athletes")
```

```python
display(df.limit(15))
```

```python
df_filter = df.filter((col('current') == True) & (col('name').isin(['GALSTYAN Slavik','HARUTYUNYAN Arsen','SEHEN Sajjad'])))

display(df_filter)
```

```python
from pyspark.sql.types import *
```

```python
df = df.withColumn('height', col('height').cast(FloatType()))\
        .withColumn('weight', col('weight').cast(FloatType()))

display(df.limit(15))
```

```python
df_sorted = df.sort('height','weight', ascending=[0,1]).filter(col('weight') > 0)

display(df_sorted.limit(15))
```

```python
df_sorted = df_sorted.withColumn('nationality_long', regexp_replace('nationality_long','United States of America','America'))

display(df_sorted.limit(15))
```

```python
df_sorted = df_sorted.withColumnRenamed('code','athlete_id')

display(df_sorted.limit(51))
```

```python
df_sorted = df_sorted.withColumn('occupation', split(col('occupation'),','))

display(df_sorted.limit(51))
```

```python
df_sorted.columns
```

```python
df_final = df_sorted.select('athlete_id',
 'current',
 'name',
 'name_short',
 'name_tv',
 'gender',
 'function',
 'country_code',
 'country',
 'country_long',
 'nationality',
 'nationality_long',
 'nationality_code',
 'height',
 'weight',
 'disciplines',
 'events')
```

```python
display(df_final.limit(10))
```

```python
display(df_final)
```

```python
df_final.write.format('delta')\
    .mode('append')\
    .option('path','abfss://silver@olympicprojdlchintu.dfs.core.windows.net/athletes')\
    .saveAsTable('olympic.silver.athletes')
```
